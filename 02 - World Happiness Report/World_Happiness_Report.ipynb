## Penjelasan Dataset

Dataset-dataset berikut ini menunjukkan score happines dunia. Score happiness menggunakan data dari GallUp world poll. Data feature yang menjadi kata kunci di report ini adalah:

- <strong>Life Ladder</strong> >>  Ibarat sebuah tangga, pijakan 0 adalah dasar dan 10 adalah atas. Tangga atas menunjukkan kehidupan terbaik dan dasar tangga menunjukkan kehidupan terburuk
- <strong>Log GDP Per capita</strong> >> nilai ekonomi yang dihasilkan setiap individu warga
- <strong>Social support</strong> >> dukungan sosial seperti keluarga, teman, tetangga dll
- <strong>Healthy life expectancy at birth</strong> >> rata-rata hidup dalam kesehatan yang baik
- <strong>Freedom to make life choices</strong> >> Peluang tiap individu untuk melakukan aksi yang dipilih
- <strong>Generosity</strong> >> Kemurahan hati
- <strong>Perceptions of corruption</strong> >> Korupsi

### Import Library and Dataset

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#import semua data

df1 = pd.read_csv('world-happiness-report.csv')
df2 = pd.read_csv('world-happiness-report-2021.csv')
```

### Statistics Summary

```python
df2.drop(['Standard error of ladder score','upperwhisker','lowerwhisker','Ladder score in Dystopia',
          'Explained by: Log GDP per capita','Explained by: Social support','Explained by: Healthy life expectancy',
         'Explained by: Freedom to make life choices','Explained by: Generosity','Explained by: Perceptions of corruption'],axis=1, inplace=True)
```

```python
df2.describe()
```

### Filtering and Cleansing Data Indonesia

```python
indonesia = df1.loc[df1['Country name'] == 'Indonesia']
indonesia.reset_index(drop=True)
```

```python
indonesia.drop(['Positive affect','Negative affect'],axis=1, inplace=True)
indonesia
```

```python
indonesia2 = df2.loc[df2['Country name'] == 'Indonesia']
indonesia2.reset_index(drop=True)
```

```python
indonesia2.drop(['Regional indicator','Dystopia + residual'],axis=1, inplace=True)
indonesia2
```

```python
indonesia2.insert(1,"year",2021,True)
```

```python
indonesia2.rename(columns={'Ladder score': 'Life Ladder',
                          'Logged GDP per capita':'Log GDP per capita',
                          'Healthy life expectancy':'Healthy life expectancy at birth'},inplace=True)
indonesia2
```

```python
data_terbaru = pd.concat([indonesia, indonesia2])
data_terbaru.reset_index(drop=True)
```

```python
data_terbaru.rename(columns={'year':'Year'},inplace=True)
data_terbaru.reset_index(drop=True)
```

### Visualization

```python
sns.set(rc={'figure.figsize':(18,6)})
sns.set_style("whitegrid")
sns.lineplot(data=data_terbaru, x="Year", y="Life Ladder", marker="o")
plt.title("Indonesia's ladder score over the years", fontweight="bold", fontsize=20)
```

### Insights:
- Angka ladder score Indonesia paling rendah pada tahun 2008, hanya 4.8
- Angka ladder score Indonesia paling tinggi pada tahun 2014, di 5.6
- Dari tahun 2018 sampai 2020, angka ladder score cenderung stabil dan tidak banyak berubah

### Setting up the Ladder Score

```python
df2[['Country name', 'Ladder score', 'Regional indicator']].sort_values(by='Ladder score',ascending=False)
```

```python
df_baru = df2.loc[df2['Country name']=='Finland']
df_baru
```

```python
df_baru.drop(['Logged GDP per capita','Social support','Healthy life expectancy','Freedom to make life choices','Generosity',
         'Perceptions of corruption','Dystopia + residual'], inplace=True, axis=1)
df_baru
```

```python
south_east_asia = df2.loc[df2['Regional indicator'] == 'Southeast Asia',  ['Country name', 'Ladder score', 'Regional indicator']]
south_east_asia
```

```python
rank_terbaru = pd.concat([df_baru, south_east_asia])
rank_terbaru
```

```python
rank_terbaru.drop(['Regional indicator'],inplace=True,axis=1)
```

```python
rank_terbaru['Rank'] = rank_terbaru.sort_values(by=['Ladder score'], ascending=False).index + 1
```
